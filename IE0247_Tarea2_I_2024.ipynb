# Serie de Fourier de una señal triangular

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal triangular
frecuencia = 1.0  # Frecuencia de la señal en Hz
amplitud = 1.0    # Amplitud de la señal

# Configuración del tiempo
tiempo_inicio = 0
tiempo_fin = 2.0  # Duración de la señal en segundos
muestreo = 1000   # Número de puntos de muestreo

# Generar la señal triangular
tiempo = np.linspace(tiempo_inicio, tiempo_fin, muestreo, endpoint=False)
señal_triangular = amplitud * np.abs(np.mod(tiempo * frecuencia, 1) - 0.5) * 2

# Gráfica de la señal triangular
plt.plot(tiempo, señal_triangular, label='Señal Triangular')
plt.title('Señal Triangular Continua')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Función para calcular la serie de Fourier para la señal triangular
def fourier_triangular(tiempo, frecuencia, n_componentes):
    suma_parcial = np.zeros_like(tiempo)
    for n in range(1, n_componentes + 1):
        componente = (8 / (np.pi**2 * n**2)) * np.sin(2 * np.pi * n * frecuencia * tiempo)
        suma_parcial += componente
    return suma_parcial

# Visualizar la convergencia de la serie de Fourier
num_componentes = 10  # Número de componentes a considerar

plt.figure(figsize=(12, 6))
plt.plot(tiempo, señal_triangular, label='Señal Triangular Original', linewidth=2)

for i in range(1, num_componentes + 1):
    serie_parcial = fourier_triangular(tiempo, frecuencia, i)
    plt.plot(tiempo, serie_parcial, label=f'{i} Componentes', linestyle='--')

plt.title('Convergencia de la Serie de Fourier para la Señal Triangular')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
