# Serie de Fourier de una se√±al triangular

```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros de la se√±al triangular
frecuencia = 1.0  # Frecuencia de la se√±al en Hz
amplitud = 1.0    # Amplitud de la se√±al

# Configuraci√≥n del tiempo
tiempo_inicio = 0
tiempo_fin = 2.0  # Duraci√≥n de la se√±al en segundos
muestreo = 1000   # N√∫mero de puntos de muestreo

# Generar la se√±al triangular
tiempo = np.linspace(tiempo_inicio, tiempo_fin, muestreo, endpoint=False)
se√±al_triangular = amplitud * np.abs(np.mod(tiempo * frecuencia, 1) - 0.5) * 2

# Gr√°fica de la se√±al triangular
plt.plot(tiempo, se√±al_triangular, label='Se√±al Triangular')
plt.title('Se√±al Triangular Continua')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Funci√≥n para calcular la serie de Fourier para la se√±al triangular
def fourier_triangular(tiempo, frecuencia, n_componentes):
    suma_parcial = np.zeros_like(tiempo)
    for n in range(1, n_componentes + 1):
        componente = (8 / (np.pi**2 * n**2)) * np.sin(2 * np.pi * n * frecuencia * tiempo)
        suma_parcial += componente
    return suma_parcial

# Visualizar la convergencia de la serie de Fourier
num_componentes = 10  # N√∫mero de componentes a considerar

plt.figure(figsize=(12, 6))
plt.plot(tiempo, se√±al_triangular, label='Se√±al Triangular Original', linewidth=2)

for i in range(1, num_componentes + 1):
    serie_parcial = fourier_triangular(tiempo, frecuencia, i)
    plt.plot(tiempo, serie_parcial, label=f'{i} Componentes', linestyle='--')

plt.title('Convergencia de la Serie de Fourier para la Se√±al Triangular')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
