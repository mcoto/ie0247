# 1) Suma de señales continuas

El siguiente códido genera dos señales senoidales continuas de diferente frecuencia y amplitud.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las señales
amplitud_1 = 1.0
frecuencia_1 = 2.0  # Hz

amplitud_2 = 0.5
frecuencia_2 = 3.0  # Hz

frecuencia_muestreo = 100.0  # Hz
tiempo_total = 1.0  # segundos

# Generar señales continuas
tiempo_continuo = np.arange(0, tiempo_total, 1/frecuencia_muestreo)
senal_1 = amplitud_1 * np.sin(2 * np.pi * frecuencia_1 * tiempo_continuo)
senal_2 = amplitud_2 * np.sin(2 * np.pi * frecuencia_2 * tiempo_continuo)

# Sumar las señales
senal_sumada = senal_1 + senal_2

# Graficar señales
plt.subplot(3, 1, 1)
plt.plot(tiempo_continuo, senal_1)
plt.title('Señal 1')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 2)
plt.plot(tiempo_continuo, senal_2)
plt.title('Señal 2')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 3)
plt.plot(tiempo_continuo, senal_sumada)
plt.title('Suma de Señales')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()

```

# 2) Suma de señales discretas

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las señales
amplitud_1 = 1.0
frecuencia_1 = 2.0  # Hz

amplitud_2 = 0.5
frecuencia_2 = 3.0  # Hz

tiempo_total = 10  # Número de muestras

# Generar señales discretas
tiempo_discreto = np.arange(0, tiempo_total)
senal_1 = amplitud_1 * np.sin(2 * np.pi * frecuencia_1 * tiempo_discreto)
senal_2 = amplitud_2 * np.sin(2 * np.pi * frecuencia_2 * tiempo_discreto)

# Sumar las señales
senal_sumada = senal_1 + senal_2

# Graficar señales
plt.subplot(3, 1, 1)
plt.stem(tiempo_discreto, senal_1)
plt.title('Señal 1')
plt.xlabel('Tiempo (muestras)')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 2)
plt.stem(tiempo_discreto, senal_2)
plt.title('Señal 2')
plt.xlabel('Tiempo (muestras)')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 3)
plt.stem(tiempo_discreto, senal_sumada)
plt.title('Suma de Señales')
plt.xlabel('Tiempo (muestras)')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

# 3) Transformaciones de una señal

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del pulso rectangular
amplitud = 1.0
ancho = 0.5
inicio = -1.0
fin = 1.0

# Generar pulso rectangular
tiempo = np.linspace(inicio, fin, num=1000)
pulso_rectangular = np.where((tiempo >= -ancho/2) & (tiempo <= ancho/2), amplitud, 0)

# Reflexiones y transformaciones
reflejada = -pulso_rectangular
desplazada = np.roll(pulso_rectangular, 100)  # Desplazamiento de 100 muestras hacia la derecha
escalada = 0.5 * pulso_rectangular
inversa = 1 - pulso_rectangular

# Configuración de la figura
fig, axs = plt.subplots(5, 1, figsize=(8, 12))

# Graficar señales
axs[0].plot(tiempo, pulso_rectangular)
axs[0].set_title('Pulso Rectangular')
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel('Amplitud')
axs[0].set_ylim([-1.5, 1.5])

axs[1].plot(tiempo, reflejada)
axs[1].set_title('Reflejada')
axs[1].set_xlabel('Tiempo')
axs[1].set_ylabel('Amplitud')
axs[1].set_ylim([-1.5, 1.5])

axs[2].plot(tiempo, desplazada)
axs[2].set_title('Desplazada')
axs[2].set_xlabel('Tiempo')
axs[2].set_ylabel('Amplitud')
axs[2].set_ylim([-1.5, 1.5])

axs[3].plot(tiempo, escalada)
axs[3].set_title('Escalada')
axs[3].set_xlabel('Tiempo')
axs[3].set_ylabel('Amplitud')
axs[3].set_ylim([-1.5, 1.5])

axs[4].plot(tiempo, inversa)
axs[4].set_title('Inversa')
axs[4].set_xlabel('Tiempo')
axs[4].set_ylabel('Amplitud')
axs[4].set_ylim([-1.5, 1.5])

plt.tight_layout()
plt.show()
```
